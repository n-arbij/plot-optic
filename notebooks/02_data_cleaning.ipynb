# 02 Data Cleaning and Visualization

### Objective

This notebook cleans raw dataset based on issues identified during exploration. The ain is to produce a reliable, analysis-readu dataser while documenting each decision made.

### Import Libraries

```python
import pandas as pd
import numpy as np
```

### Load Raw Data

```python
df = pd.read_csv("../data/raw/dataset.csv")
df.head()
```

### Standardize Columns Names

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

df.columns
```

Reasoning:
Standardized column names improve readability and reduce errors when referencing columns.

### Remove Duplicate Records

```python
duplicates = df.duplicated().sum()
duplicates

df = df.drop_duplicates()
```

Reasoning: Duplicate records can bias analysis and must be removed.

### Convert Data Types

##### Convert Data Columns

```python
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

Resoning: Invalid data values are coerced to NaT for later handling.

##### Convert Numeric Columns Stored as Text

```python
df['amount'] = pd.to_numeric(df('amount'), errors='coerce')
```

Reasoning: Ensures numric operations can be performed safely.

### Handle Missing Values

##### Check Missing Values

```python
df.isnull().sum()
```

Stategies:

- Critical columns: remove rowa with missing values
- Non-critical columns: impute or leave as-is depending on impact


##### Drop Rows with Missing Critical Values

```python
df = df.dropna(subset=['date', 'amount'])
```

Reasoning: Preserves records while clearing marking missing categories.

##### Fill Missing Categorical Values

```python
df['category'] = df['category'].fillna('unknown')
```

Reasoning: Preserve records while clearly marking missing categories.

### Standardize Categorical Values

```python
df['category'] = df['category'].str.lower().str.strip()
```

Reasoning: Reduce duplicate categories caused by inconsistent formatting.

### Outlier Check (Basic)

```python
df['amount'].describe()
```

Observation: Extreme values exist and will be handled during analysis if necessary.

### Final Data Validation

```python
df.info()

df.head()
```

Validation Checks:
- Correct data types
- Reduced missing values
- No duplicate records

### Save Clean Dataset

```python
df.to_csv("../data/processed/clean_dataset.csv", index=False)
```

Reasoning: Saving the cleaned dataset ensures reproducibility and separates row from processed data.

### Conclusion

The dataset is now clean and ready for analysis and visualization. All cleaning steps were driven by issues observed during exploration.
